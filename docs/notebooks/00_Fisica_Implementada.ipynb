# 0. Física Implementada en PA3Py

Este notebook contiene la documentación científica completa del modelo físico subyacente de `PA3Py`.

## 1. Acreción de Pebbles (Pebble Accretion)

El núcleo del simulador resuelve el crecimiento del núcleo planetario gobernado por la acreción de *pebbles* (Lambrechts & Johansen 2012, 2014). Dependiendo de si la capa de *pebbles* es más gruesa o más delgada que el radio de Hill del embrión, el régimen puede ser 2D o 3D.

### Régimen 3D

Cuando el embrión es pequeño, su radio de acreción $R_{acc}$ es menor que la escala de altura de los *pebbles* ($H_d$). La tasa de acreción volumétrica es:

$$ \dot{M}_{3D} = \pi R_{acc}^2 \Delta v \rho_{d} $$

Donde $\rho_d = \frac{\Sigma_d}{\sqrt{2\pi} H_d}$ es la densidad volumétrica de los *pebbles* en el plano medio.

### Régimen 2D

A medida que el embrión crece y su $R_{acc}$ excede $H_d$, la acreción se vuelve puramente bidimensional (disco aplanado desde la perspectiva del planeta):

$$ \dot{M}_{2D} = 2 R_{acc} \Sigma_d \Delta v $$

El simulador interpola suavemente entre ambos regímenes.

## 2. Masa de Transición ($M_t$)

La eficiencia de acreción depende del parámetro aerodinámico de Stokes ($St$). Cuando el número de Stokes excede $0.1$, el régimen entra en un límite de acreción óptima.

La masa crítica donde la acreción es máxima para un $St$ dado es la Masa de Transición (Bitsch et al. 2015):

$$ M_t = \sqrt{\frac{1}{3}} \frac{\Delta v^3}{G \Omega} $$

## 3. Masa de Aislamiento de Pebbles ($M_{iso}$)

A medida que el planeta gana masa, perturba el gas a su alrededor creando un gradiente de presión positivo (un *pressure bump*) fuera de su órbita. Este gradiente atrapa los *pebbles* que derivan hacia adentro, cortando el suministro.

`PA3Py` utiliza la prescripción de masa de aislamiento definida por Bitsch et al. (2018):

$$ M_{iso} = 25 M_\oplus \left( \frac{H/r}{0.05} \right)^3 \left( \frac{a}{1 \times 10^{-3}} \right) \times \Pi $$

Donde $\Pi$ es una función compleja de los parámetros del disco y el gradiente de presión local.

## 4. Evolución de la Snowline y Química

Las *snowlines* dictan qué especies se condensan sobre los *pebbles*. En el modelo por defecto, la posición de la línea de hielo de agua ($r_{snow}$) evoluciona en el tiempo debido a los cambios de temperatura del disco (radiación pasiva + calentamiento viscoso), de acuerdo a los tratamientos teóricos tipo Oka et al. (2011) o extrayendo la temperatura directamente de simulaciones numéricas.
