# Poblaciones Sintéticas y Diagramas de Hovmöller
Este notebook enseña cómo simular muchos embriones simultáneamente y usar los gráficos integrados de `PA3Py`.


```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

sys.path.insert(0, os.path.abspath('../../src'))
from pa3py import PA3Py
from pa3py import constants as c
```

## 1. El Disco (Diagrama de Hovmöller)
`PA3Py` trae gráficos integrados. Solo llama a `plot_hovmoller`.


```python
data_path = '../../tests/test_data/run_smooth_a0.001_v10'
sim = PA3Py(data_path)

# Graficamos la densidad de superficie del polvo
fig, ax = sim.plot_hovmoller(field='dust_Sigma', show_snowlines=True)
plt.show()
```

## 2. Simulando una Población de 100 Embriones


```python
# 100 embriones desde 1 AU hasta 30 AU
embryos_au = np.linspace(1.0, 30.0, 100).tolist()

# Corremos la simulación de toda la población (¡PA3Py es muy rápido!)
resultados = sim.run_growth(embryos_au, m_seed_me=1e-3)
```

## 3. Análisis Poblacional: Masa Final vs Masa de Aislamiento


```python
# Extraemos la masa de aislamiento del objeto central
m_iso_map = sim.calculate_isolation_mass_map()
m_iso_t0 = m_iso_map[0, :] / c.M_EARTH
r_disco = sim.disk.r / c.AU

# Extraer masas finales de la población
radios = []
masas_finales = []

for r_au, hist in resultados.items():
    if len(hist) > 0:
        radios.append(r_au)
        masas_finales.append(hist[-1][1] / c.M_EARTH)

plt.figure(figsize=(10, 5))
plt.plot(r_disco, m_iso_t0, 'k--', alpha=0.5, label='Isolation Mass teórica (t=0)')
plt.scatter(radios, masas_finales, c=radios, cmap='viridis', s=50, edgecolor='k', label='Masa Final Alcanzada')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Radio Orbital [AU]')
plt.ylabel('Masa del Planeta [$M_\\oplus$]')
plt.title('Población Sintética: Masa Final vs Distancia')
plt.xlim(0.5, 50)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
