# PA3Py Quickstart: Crecimiento de Waterworlds
Este notebook muestra la forma más sencilla de correr PA3Py usando la interfaz unificada `PA3Py`.


```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

# Añadir PA3Py al path local
sys.path.insert(0, os.path.abspath('../../src'))
from pa3py import PA3Py
```

## 1. Inicializar la Simulación
Solo necesitamos la ruta a un archivo HDF5 válido de TripodPy/DustPy.


```python
# NOTA: Asegúrate de apuntar a la ruta correcta donde tienes tus datos
data_path = '../../tests/test_data/run_smooth_a0.001_v10'

# Un solo objeto centraliza todo el paquete
sim = PA3Py(data_path)
```

## 2. Correr la Acreción
Corremos la simulación para embriones en 1, 5 y 15 AU.


```python
resultados = sim.run_growth(embryos_au=[1.0, 5.0, 15.0], m_seed_me=1e-3)
```

## 3. Visualización
Graficaremos la masa total vs el tiempo para cada planeta.


```python
plt.figure(figsize=(10, 6))

for r_au, hist in resultados.items():
    if len(hist) == 0: continue
    
    tiempos_myr = hist[:, 0] / (1e6 * 3.15576e7)  # Convertir seg a Myr
    masa_total = hist[:, 1] / 5.97e27             # Convertir g a Masas Terrestres
    
    plt.plot(tiempos_myr, masa_total, label=f'Embrión en {r_au} AU')

plt.xlabel('Tiempo [Myr]')
plt.ylabel('Masa del Núcleo [$M_\\oplus$]')
plt.yscale('log')
plt.legend()
plt.grid(True, alpha=0.3)
plt.title('Crecimiento de Embriones Planetarios')
plt.show()
```
